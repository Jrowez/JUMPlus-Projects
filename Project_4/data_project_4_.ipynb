```python
"""2. Import required libraries"""
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
"""9. Group the sales data of each region on Year_of_Release and plot the same
(It should show which region sales was at peak on which year of release)"""

release_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Year_of_Release']]
release_sales = release_sales.groupby('Year_of_Release').sum()
release_sales

figsize = (15, 8)
fig, ax = plt.subplots(figsize=figsize)
genre = sns.lineplot(data=release_sales, ax=ax, )


```

```python
"""10. Group North American, Japan, European & Other region sales data as per Genre and plot it (It should show highest of region according to game
category , ex- NA_Sales is highest in Adventure category , plot kind should be bar)"""

genre_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Genre']]
genre_sales = genre_sales.groupby('Genre', as_index=False).sum()
genre_sales

fig, ax =plt.subplots(figsize=(15,8))
genre_NA = sns.barplot(x='Genre', y='NA_Sales', data=genre_sales, color='Red')
genre_EU = sns.barplot(x='Genre', y='EU_Sales', data=genre_sales, color='Blue')
genre_JP = sns.barplot(x='Genre', y='JP_Sales', data=genre_sales, color='Green')
genre_Other = sns.barplot(x='Genre', y='Other_Sales', data=genre_sales, color='Purple',)
```

```python
"""11. Group North American, Japan, European & Other region sales data as per Platform and plot 
it (Should show sales as per "Platform genre" , plot kind should be bar )"""

platform_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Platform']]
platform_sales = platform_sales.groupby('Platform', as_index=False).sum()
platform_sales

fig, ax =plt.subplots(figsize=(15,8))
platform_NA = sns.barplot(x='Platform', y='NA_Sales', data=platform_sales, color='Red')
platform_EU = sns.barplot(x='Platform', y='EU_Sales', data=platform_sales, color='Blue')
platform_JP = sns.barplot(x='Platform', y='JP_Sales', data=platform_sales, color='Green')
platform_Other = sns.barplot(x='Platform', y='Other_Sales', data=platform_sales, color='Purple',)
```

```python
"""12. Plot the Sales in all regions per Platform , you need to create a boxplot for that which will show the distribution of sales as per platform
"""
#TODO: all of this
total_region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Platform', 'Total_Sales']]
total_region_sales = total_region_sales.groupby('Platform', as_index=False).sum()
total_region_sales
figsize = (15, 8)
fig, ax = plt.subplots(figsize=figsize)
genre = sns.boxplot( x='Platform',  ,data=total_region_sales)
```

```python
"""13. Find out which Genre games are sold in highest numbers (create a countplot of seaborn which could do that) """
popular_genres = df[['Genre', 'Total_Sales']]
popular_genres = popular_genres.groupby('Genre').sum()
popular_genres

popular_genres['Genre']= popular_genres.index
plt.figure(figsize=(8,4))
popular_genre_plot = sns.catplot(x='Genre', y='Total_Sales', data=popular_genres, kind='bar')
popular_genre_plot.set_xticklabels(['Action', 'Adventure', 'Fighting', 'Misc', 'Platform', 'Puzzle', 'Racing', 'Role-POlaying', 'Shooting', 'Simulation', ''], rotation=90)
plt.show()
```

```python
"""14. Try to find out global sales as per Genre ,(create a barplot of seaborn for that) """
sales_per_genre = df[['Genre', 'Global_Sales']].groupby('Genre', as_index=False).sum().sort_values('Global_Sales', ascending=False)
sales_per_genre

fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(x='Genre', y='Global_Sales', data=sales_per_genre, ax=ax)

```
