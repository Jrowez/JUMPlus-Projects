```python
"""1. Import required libraries"""

import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt 
import numpy as np
```

```python
"""2. Read the csv file"""
raw_df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
raw_df
```

```python
"""3. Show first five results"""
raw_df.head()
```

```python
"""4. Show if columns have any NaN,N/A or null values in boolean form"""

#cleaner version
raw_df.isna().any()

#table format
raw_df.isna()


```

```python
"""5. Check how many null,empty,N/A values are there in each column"""
raw_df.isnull().sum()

```

```python
raw_df.isna().sum()
```

```python
"""6. Show null values of each column with the help of Heatmap"""
figsize = (10, 5)
fig, ax = plt.subplots(figsize=figsize)
# sns.heatmap(data=raw_df.isna().any())
sns.heatmap(raw_df.isna(), yticklabels=False)
```

```python
"""7. Drop outliers (columns which are not useful according to you)"""

#dropping Critic_Score, Critic_Count, User_Score, User_Count, Developer
df = raw_df.drop(['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer'], axis=1)
```

```python
df.head()
```

```python
"""8. Drop the null values containing rows from the object which you created in point 7"""
clean_df = df.dropna()
```

```python
"""9. Check the dataframe again if there is anything which you want to remove"""
clean_df
```

```python
"""10. Export dataframe to your desired location with no index column"""

clean_df.to_csv('Video_Games_Sales_22_Dec_2016_Clean', index=False)
```
